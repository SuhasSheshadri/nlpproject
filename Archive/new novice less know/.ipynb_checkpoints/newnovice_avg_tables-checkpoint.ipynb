```python
import pandas as pd
import numpy as np
```

```python
def read_expert(filename):
    data = pd.read_csv(filename, sep=":", header=None)
    data.columns = ["word", "expert_score"]
    data.set_index('word',inplace=True)
    return data
```

```python
def read_novice(filename):
    data = pd.read_csv(filename, sep=":", header=None)
    data.columns = ["word", "novice_score"]
    data.set_index('word',inplace=True)
    return data
```

```python
def read_verynovice(filename):
    data = pd.read_csv(filename, sep=":", header=None)
    data.columns = ["word", "very_novice_score"]
    data.set_index('word',inplace=True)
    return data
```

```python
def cal_average(novice_file,expert_file,verynovice_file):
    d1 = read_expert(expert_file)
    d2 = read_novice(novice_file)
    d3 = read_verynovice(verynovice_file)
    d4 = d1.join(d2)
    d5 = d4.join(d3) 
    d5['expert-novice'] = d5['expert_score']-d5['novice_score']
    d5['expert-verynovice'] = d5['expert_score']-d5['very_novice_score']
    d5.loc['Average'] = d5.mean()
    return d5
```

```python
path='/Users/Rithika/nlpproject/Archive/cosine_sim_new_describe_words/'
p = cal_average(path+'avg_badminton_novice.txt',path+'avg_badminton_expert.txt','avg_badminton_newnovice.txt')
print(p)
#p.to_html('badminton.html')
```
