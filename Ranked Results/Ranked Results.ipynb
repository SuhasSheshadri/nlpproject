```python
import ast
import codecs
import os

# Uncomment the files to be used. Comment the remaining ones
# file = "Ranked Results - Badminton.txt"
# nov = codecs.open("Top_200/badminton_novice.txt", 'r',"utf-8")
# exp = codecs.open("Top_200/badminton_expert.txt", 'r',"utf-8")

# file = "Ranked Results - Dance.txt"
# nov = codecs.open("Top_200/dance_novice.txt", 'r',"utf-8")
# exp = codecs.open("Top_200/dance_expert.txt", 'r',"utf-8")

# file = "Ranked Results - Shooter.txt"
# nov = codecs.open("Top_200/shooter_novice.txt", 'r',"utf-8")
# exp = codecs.open("Top_200/shooter_expert.txt", 'r',"utf-8")

file = "Ranked Results - Psychotherapy.txt"
nov = codecs.open("Top_200/psychotherapy_novice.txt", 'r',"utf-8")
exp = codecs.open("Top_200/psychotherapy_expert.txt", 'r',"utf-8")

try:
    os.remove(file)
except OSError:
    pass


g = codecs.open(file, "a", "utf-8")

for line in nov:
    nov_nxt_line = line
    
    exp_nxt_line = exp.readline()

    nov_similaritys = []
    nov_words,nov_similaritys = nov_nxt_line.split(":")
    
    exp_similaritys = []
    exp_words,exp_similaritys = exp_nxt_line.split(":")
    
    nov_similaritys = ast.literal_eval(nov_similaritys)
    a,b = [list(c) for c in zip(*nov_similaritys)]

    exp_similaritys = ast.literal_eval(exp_similaritys)
    c,d = [list(c) for c in zip(*exp_similaritys)]
    print("For the word '{}', the nearest words are: ".format(nov_words))
    print("\nNovice List: ")
    print(a)
    print("\nExpert List: ")
    print(c)
    print("\nComparing the words between Novice and Expert list, we can observe: ")
    
    g.write("For the word '{}', the nearest words are: \n".format(nov_words))
    g.write("\nNovice List (Top 20): \n")
    g.write(str(a))
    g.write("\n\nExpert List (Top 100): \n")
    g.write(str(c))
    g.write("\n\nComparing the words between Novice and Expert list, we can observe: \n")
    
    i = 1
    delta = 0
    new = 0
    for word in a:
        
        try:
            exp_pos = c.index(word)+1
            delta = abs(exp_pos - i)
            
            if delta == 0:
                print("\nWord '{}' found at position {} in both lists".format(word,i,exp_pos))
                g.write("\nWord '{}' found at position {} in both lists".format(word,i,exp_pos))
                
            else:
                print("\nWord '{}' found at position {} and {}".format(word,i,exp_pos))
                g.write("\nWord '{}' found at position {} and {}".format(word,i,exp_pos))
                print("Difference between the positions: {}".format(delta))
                g.write("\nDifference between the positions: {}".format(delta))
                new += 1 
            
        except ValueError:
            print("\n'{}' at position {} is not present in Expert list".format(word, i))
            g.write("\n'{}' at position {} is not present in Expert list".format(word, i))
        i = i+1
        g.write("\n")
    print("\n")
    g.write("\n\n")
    
g.close()
nov.close()
exp.close()
```
